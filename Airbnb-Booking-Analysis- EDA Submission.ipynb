### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

df2 = Airbnb.copy()
````

````python
# Checking the number of rows with zero price values in our dataset.
df2[df2.price==0].shape
````

````python

# Removing rows having zero price values from our dataset.
df2 = df2[df2['price'] !=0]
````

````python

# Since both 'name' and 'host_name' has very less missing values as compared to the
# Entire dataset we can fill these missing values in both with unknown & no_name respectively!

df2['name'].fillna('unknown',inplace=True)
df2['host_name'].fillna('anonymous',inplace=True)
````

````python
# Checks if any null values still exist!
df2[['host_name','name']].isnull().values.any()
````

````python

# Filling null values in reviews_per_month column with 0
df2['reviews_per_month'] = Airbnb['reviews_per_month'].fillna(0)
````

````python

# Drop the last_review column
df2.drop(['last_review'], axis=1, inplace=True)
````

````python

# Checking change in shape
df2.shape
````

````python

# Checking the changes
df2.isnull().sum()
````

````python

# Boxplot of 'price' column
sns.boxplot(df2['price'], color='#cafe48')
````

Removing Outliers using IQR approach! (has performed best in removing outliers)



````python
# Calculate the IQR and boundaries
Q1 = df2['price'].quantile(0.25)
Q3 = df2['price'].quantile(0.75)
IQR = Q3 - Q1
lower_boundary = Q1 - 1.5 * IQR
upper_boundary = Q3 + 1.5 * IQR

# Filter the dataset to remove outliers
df2 = df2[(df2['price'] >= lower_boundary) & (df2['price'] <= upper_boundary)]
````

````python
#the boxplot for price now looks good without outliers!
sns.boxplot(df2['price'], color='#cafe48')
````

````python

# Distribution Plot of 'price' column
sns.distplot(df2['price'], color='#f05d5e')
````

Outliers in the minimum nights column



````python
# Calculate the IQR and boundaries
Q1 = df2['minimum_nights'].quantile(0.25)
Q3 = df2['minimum_nights'].quantile(0.75)
IQR = Q3 - Q1
lower_boundary = Q1 - 1.5 * IQR
upper_boundary = Q3 + 1.5 * IQR

# Filter the dataset to remove outliers
df2 = df2[(df2['minimum_nights'] >= lower_boundary) & (df2['minimum_nights'] <= upper_boundary)]
````

````python
# Boxplot of 'minimum_nights' column without outliers
sns.boxplot(df2['minimum_nights'], color='#cafe48')

# The boxplot for minimum nights now looks good without outliers!
````

````python
#Lets check the statistical value(mean, median) of all feature of neighbourbood group wise

df_mean_median = df2[['neighbourhood_group','price','minimum_nights','number_of_reviews','availability_365']]
#defining a function get_mean_median_mode to fing mean and median
def get_mean_median(data,neighbourhood_group):
  ''' This fuction gives mean and median of all features'''
  # try block to utilize eror handling
  try:
    return pd.concat([df_mean_median[df_mean_median['neighbourhood_group']==neighbourhood_group].describe().round(2).iloc[1],
          df_mean_median[df_mean_median['neighbourhood_group']==neighbourhood_group].describe().round(2).iloc[5]], axis = 1).rename(columns={'50%':'Median'})
  # except block to catch error
  except:
    print("Invalid Neighbourhood Group")
````

````python

# using a for loop to print the mean and median for all unique values from the neighborhood group
for i in Airbnb.neighbourhood_group.unique():
  print(i,get_mean_median(df_mean_median,i))
  print('\n')
````

#Chart - 1 Column wise Histogram and Box Plot


````python
# Chart - 1 visualization code
for col in df2.describe().columns:
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

    # Histogram plot
    sns.histplot(df2[col], ax=axes[0], kde=True, color='#05204a', edgecolor='white')
    axes[0].set_xlabel(col)
    axes[0].set_ylabel('Frequency')
    axes[0].set_title("Distribution of " + col, fontsize=15)
    axes[0].grid(True, linestyle='--')

    # Boxplot
    sns.boxplot(df2[col], ax=axes[1], orient='h', showmeans=True, color='#cafe48')
    axes[1].set_xlabel(col)
    axes[1].set_ylabel('')
    axes[1].set_title("Boxplot of " + col, fontsize=15)
    axes[1].grid(True, linestyle='--')

    # Adjust spacing between subplots
    plt.tight_layout()

    # Show the plots
    plt.show()
````

##### 1. Why did you pick the specific chart?

It is also useful when dealing with large data sets (greater than 100 observations). It can help detect any unusual observations (outliers) or any gaps in the data.

Thus, we used the histogram plot to analyse the variable distributions over the whole dataset whether it's symmetric or not.

A boxplot is used to summarize the key statistical characteristics of a dataset, including the median, quartiles, and range, in a single plot. Boxplots are useful for identifying the presence of outliers in a dataset, comparing the distribution of multiple datasets, and understanding the dispersion of the data. Thus, for each numerical varibale in the given dataset, I used box plot to analyse the outliers and interquartile range including median, maximum and minimum value.

````python
# Chart - 2 visualization code
# Chart - 2 visualization code
# Group the dataset by "Neighbourhood group" and calculate the average price
neighborhood_avg_price = df2.groupby('neighbourhood_group')['price'].mean()

# Set the style of the plot
sns.set(style="whitegrid")

# Create a bar plot with custom color palette
plt.figure(figsize=(10, 6))
colors = ['#272932', '#f05d5e', '#0f7173', '#9fc2cc', '#dbc648']
ax = sns.barplot(x=neighborhood_avg_price.index, y=neighborhood_avg_price, palette=colors)

# Set the title and axis labels
plt.title('Neighbourhood Group vs. Average Price', fontsize=16)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=30)

# Add value labels on top of each bar
for p in ax.patches:
    ax.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', xytext=(0, 5), textcoords='offset points', fontsize=10)

# Remove the top and right spines
sns.despine()

#display the plot
plt.tight_layout()
plt.show()
````

#### Chart - 5

What is the average price of Airbnb listings in NYC in 2019, categorized by room type?



````python
# Chart - 5 visualization code
# Select the relevant columns
df = df2[['neighbourhood_group', 'room_type', 'price']]

# Group the data by neighbourhood_group and room_type, and calculate the mean price
df = df.groupby(['neighbourhood_group', 'room_type'], as_index=False)[['price']].mean()

# Create a figure and set the theme to white
plt.figure(figsize=(12, 6))
sns.set_theme(style='white')

# creating palette
my_pal = {'Entire home/apt': '#f05f5e', 'Private room': '#ead94c', 'Shared room': '#0f7173'}

# Create a bar plot with x as neighbourhood_group, y as price, hue as room_type, and using the custom color palette
df = sns.barplot(x="neighbourhood_group", y="price", data=df, hue='room_type', palette=my_pal)

# Add annotations to the bars with the formatted average price values
for p in df.patches:
    df.annotate(format(p.get_height(), '.2f'), (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', size=11, xytext=(0, -12), textcoords='offset points')

# Set the x and y labels, and the title of the plot
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.title('Average Price by Room Type')

# Show the plot
plt.show()
````
